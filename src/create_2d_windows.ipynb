```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
def draw_bboxes_with_dimensions(bboxes_dict):
    """
    Draw multiple bounding boxes with their names on a 2D plane of HD size (1920x1080 pixels), using a dictionary,
    where each bbox is specified by its top-left corner and its dimensions (width and height).

    Parameters:
    bboxes_dict: A dictionary where each key is the name of a bbox and the value is a tuple containing the coordinates
                 of the top-left corner and the dimensions (width, height) of the bbox.
                 Format: {name: (x, y, w, h)}
    """

    # Create a figure and axis with HD dimensions
    fig, ax = plt.subplots(figsize=(19.20, 10.80))

    # Iterate over the dictionary of bbox names and dimensions, and add each as a rectangle patch
    for name, (x, y, w, h) in bboxes_dict.items():
        rect = patches.Rectangle((x, y), w, h, linewidth=1, edgecolor='r', facecolor='none')
        ax.add_patch(rect)

        # Add the name of the bbox
        ax.text(x, y, name, verticalalignment='top', color='blue', fontsize=12, weight='bold')

    # Set the limits of the axis
    ax.set_xlim(0, 1920)
    ax.set_ylim(0, 1080)
    ax.set_aspect('equal', adjustable='box')

    # Display the plot
    plt.gca().invert_yaxis()
    plt.show()
```

```python
# Example usage with bbox dictionary and dimensions
bbox_dict_with_dimensions = {
    "Box1": (100, 100, 400, 300),  # x, y, w, h
    "Box2": (600, 200, 600, 600),
    "Box3": (1300, 50, 600, 250),
    "Box4": (1250, 400, 600, 600),
}
draw_bboxes_with_dimensions(bbox_dict_with_dimensions)
```
